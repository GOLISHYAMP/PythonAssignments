### Assignment 6: Time Series Analysis

1. Create a Pandas DataFrame with a datetime index and one column filled with random integers. Resample the DataFrame to compute the monthly mean of the values.
2. Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31' and one column filled with random integers. Compute the rolling mean with a window of 7 days.

```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Function to generate a random date within a range
def random_date(start, end):
    delta = end - start
    random_days = random.randint(0, delta.days)
    return start + timedelta(days=random_days)

# Define the start and end date range
start_date = datetime(2025, 1, 1)
end_date = datetime(2025, 12, 31)

# Generate a DataFrame with 10 rows of random dates
df = pd.DataFrame({
    'random_dates': [random_date(start_date, end_date) for _ in range(100)],
    'Sales' : np.random.randint(60, 100, 100)
})
df.set_index('random_dates', inplace=True)
df.head()
```

```python
df.resample('ME').mean()
```

```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Function to generate a random date within a range
def random_date(start, end):
    delta = end - start
    random_days = random.randint(0, delta.days)
    return start + timedelta(days=random_days)

# Define the start and end date range
start_date = datetime(2021, 1, 1)
end_date = datetime(2024, 12, 31)

# Generate a DataFrame with 10 rows of random dates
df = pd.DataFrame({
    'date': [random_date(start_date, end_date) for _ in range(1000)],
    'values' : np.random.randint(60, 100, 1000)
})
# df.set_index('date', inplace=True)
df.head()
```

```python
df['quater'] = df['date'].dt.quarter
df['year'] = df['date'].dt.year
```

```python
df['quater'] = df['quater'].map({1 : 'Q1', 2: 'Q2', 3 : 'Q3', 4 : 'Q4'})
df.head()
```

```python
df.head()
```

```python
pd.pivot_table(data = df, values='values' , index = 'year', columns='quater')
```
